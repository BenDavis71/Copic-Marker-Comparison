```python
import numpy as np
import pandas as pd
from skimage import color
```

```python
#read in copic marker -> hex table
df = pd.read_html('https://github.com/mihaelamj/uicolor-copic')[0]
df
```

```python
#remove uneccesary info and rename columns
df = df[[1,3]]
df.columns = ['marker_name', 'hex']
df
```

```python
#keep only the relevant portion of the marker name
df['marker_name'] = df['marker_name'].apply(lambda x: x.split('_')[1])
df
```

```python
#convert hex to rgb
#code stole from https://stackoverflow.com/questions/29643352/converting-hex-to-rgb-value-in-python
df['rgb'] = df['hex'].apply(lambda x: list(int(x.lstrip('#')[i:i+2], 16) for i in (0, 2, 4)))
df
```

```python
#convert rgb values to cielab
#have to scale rgb values from 0-1 because skimage requires it 
df['lab'] = df['rgb'].apply(lambda x: color.rgb2lab([v/255 for v in x]))
df
```

```python
#write to parquet file
df.to_parquet('copic_colors.parquet')
```
